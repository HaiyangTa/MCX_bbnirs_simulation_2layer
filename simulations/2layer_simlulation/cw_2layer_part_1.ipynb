### Read CSV and coef

```python
# csv file to simulate: change the path!.
csv_path = '/content/drive/MyDrive/MCX_data/csv/parameter_combinations_part_1.csv'
df = pd.read_csv(csv_path)
# simulations save path
save_path = '/content/drive/MyDrive/MCX_data/result_folder/'
```

```python
lambdas = [784, 800, 818, 835, 851, 868, 881, 894]
coef_path = '/content/drive/MyDrive/MCX_data/EC.xlsx'
extinction_coeffs = pd.read_excel(coef_path)
```

```python
part_str = csv_path.split('/')[-1].replace('.csv', '').split('_')[-2] + '_' + csv_path.split('_')[-1].replace('.csv', '')
print(part_str)
```

### MCX Simulation using packgae

```python
# Prepare column names just once
col_names = ['ID']
for dist in sorted([20, 25, 30, 35]):
    for wl in lambdas:
        col_names.append(f"{dist}_{wl}")
```

```python
results_batch = []

for idx, row in df.iterrows():
    if row['complete']:
        continue

    HbO1 = row['HBO1']
    Hb1 = row['HHB1']
    HbO2 = row['HBO2']
    Hb2 = row['HHB2']
    l1 = row['Thickness']
    ID = row['ID']

    # Run simulation
    result_dict = ms2(HbO1, Hb1, HbO2, Hb2, l1, extinction_coeffs,
                      tend=1e-08, devf=1, nphoton=5e7, source_type='laser')
    df.at[idx, 'complete'] = True

    # Flatten result_dict into one row with ID
    flat_row = [ID]
    for dist in sorted(result_dict.keys()):  # 20, 25, 30, 35
        for i, wl in enumerate(lambdas):     # 8 wavelengths
            flat_row.append(result_dict[dist][i])

    # flat_row : [ID, ...] -> 1 X 33 vector
    results_batch.append(flat_row) # N x 1 x 33 list

    # Every 5 runs, save a CSV
    if len(results_batch) >= 5:
        out_path = save_path + f"{part_str}_{results_batch[-1][0]-5+1}_{results_batch[-1][0]}.csv"
        pd.DataFrame(results_batch, columns=col_names).to_csv(out_path, index=False)
        print(f"✅ Saved {len(results_batch)} results to: {out_path}")
        results_batch = []  # reset
        # mark as True:
        df.to_csv(csv_path, index=False)

# save remaining
if results_batch:
    last_ID = results_batch[-1][0]
    out_path = save_path + f"{part_str}_{last_ID}.csv"
    pd.DataFrame(results_batch, columns=col_names).to_csv(out_path, index=False)
    print(f"✅ Saved remaining {len(results_batch)} results to: {out_path}")
```
