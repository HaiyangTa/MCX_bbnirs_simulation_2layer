```python
import pandas as pd
import glob
import os
```

### Simulation data

```python
csv_folder = 'C:/Users/Haiya/2layer_sim/MCX_bbnirs_simulation_2layer/simulations/2layer_simlulation/sim_results/1_4/'
GT_folder = 'C:/Users/Haiya/2layer_sim/MCX_bbnirs_simulation_2layer/simulations/2layer_simlulation/data_generation/split_files/'
```

### Merge data

```python
GT_all['ID'] = GT_all['ID'].astype(int)
df_all['ID'] = df_all['ID'].astype(int)
GT_all = GT_all.sort_values('ID').reset_index(drop=True)
df_all = df_all.sort_values('ID').reset_index(drop=True)
```

```python
merged_df = pd.merge(df_all, GT_all, on='ID', how='inner')
```

```python
merged_df.head
```

```python
X = merged_df[df_all.columns.difference(['ID'])]
Y = merged_df[['HBO1', 'HHB1', 'HBO2', 'HHB2', 'Thickness']]
```

```python
X
```

```python
Y
```

```python
X = X.to_numpy()
Y = Y.to_numpy()
```

### ANN based on pytorch

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
```

```python
# Split into train/test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)

# Custom PyTorch dataset
class MCXSimDataset(Dataset):
    def __init__(self, X, Y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.Y = torch.tensor(Y, dtype=torch.float32)
    
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        return self.X[idx], self.Y[idx]

# Create DataLoaders
train_loader = DataLoader(MCXSimDataset(X_train, Y_train), batch_size=1, shuffle=True)
test_loader = DataLoader(MCXSimDataset(X_test, Y_test), batch_size=1, shuffle=False)
```

```python
class ANN(nn.Module):
    def __init__(self):
        super(ANN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(32, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 5)  # 5 outputs: HBO1, HHB1, HBO2, HHB2, Thickness
        )
    
    def forward(self, x):
        return self.net(x)
model = ANN()
```

### Model training

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
num_epochs = 100

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for batch_X, batch_Y in train_loader:
        optimizer.zero_grad()
        outputs = model(batch_X)
        loss = criterion(outputs, batch_Y)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss / len(train_loader):.4f}")
```

```python
model.eval()
with torch.no_grad():
    total_mse = 0
    for batch_X, batch_Y in test_loader:
        preds = model(batch_X)
        print('preds=', preds)
        print('GT=', batch_Y)
        mse = criterion(preds, batch_Y)
        total_mse += mse.item()
    print(f"Test MSE: {total_mse / len(test_loader):.4f}")
```
